```python
positive=0
negative=0
neutral=0

for comment in comments:
    sentiment=analyzer.polarity_scores(comment)
    if sentiment['compound'] >= 0.05:
        positive+=1
    elif sentiment['compound'] <= -0.05:
        negative+=1
    else:
        neutral+=1

print("Positive : ",positive)
print("Negative : ",negative)
print("Neutral : ",neutral)
```

```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import string
import re

# nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    processed_text = []
    for t in text:
        t = t.lower()  # Convert to lowercase
        t = re.sub(r'[^\w\s]', '', t)  # Remove punctuation
        processed_text.append(t)
    return processed_text

# List of your comments

# Preprocess the comments
processed_comments = preprocess_text(comments)

# Initialize the CountVectorizer to count the word frequencies
vectorizer = CountVectorizer(stop_words='english')  # Remove stop words automatically
X = vectorizer.fit_transform(processed_comments)

# Get the feature names (terms)
terms = vectorizer.get_feature_names_out()

# Get the word frequencies for each term in all documents
word_frequencies = X.toarray().sum(axis=0)

# Create a dictionary with terms and their corresponding frequencies
frequency_dict = {term: word_frequencies[i] for i, term in enumerate(terms)}

# Sort the dictionary based on frequency (higher frequency = more important)
sorted_frequency = sorted(frequency_dict.items(), key=lambda x: x[1], reverse=True)

# Display top 15 most frequent words
print("Top 50 Most Frequent Words in the Entire Document:")
for term, freq in sorted_frequency[:50]:
    print(f"{term}: {freq}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import re
from nltk.corpus import stopwords


# Define stopwords explicitly if you want to customize
stop_words = stopwords.words('english')

def preprocess_text(text):
    processed_text = []
    for t in text:
        t = t.lower()  # Convert to lowercase
        t = re.sub(r'[^\w\s]', '', t)  # Remove punctuation
        # Handle special cases like "plz" -> "please", etc.
        t = re.sub(r'\b(plz|pls|mr|beast)\b', '', t)  # Remove words like 'plz', 'pls', 'mr', 'beast'
        processed_text.append(t)
    return processed_text

# Preprocess the comments
processed_comments = preprocess_text(comments)

# Initialize the CountVectorizer to count the word frequencies, excluding stop words
vectorizer = CountVectorizer(stop_words=stop_words, ngram_range=(1, 2))  # Consider bigrams too
X = vectorizer.fit_transform(processed_comments)

# Get the feature names (terms)
terms = vectorizer.get_feature_names_out()

# Get the word frequencies for each term in all documents
word_frequencies = X.toarray().sum(axis=0)

# Create a dictionary with terms and their corresponding frequencies
frequency_dict = {term: word_frequencies[i] for i, term in enumerate(terms)}

# Sort the dictionary based on frequency (higher frequency = more important)
sorted_frequency = sorted(frequency_dict.items(), key=lambda x: x[1], reverse=True)

# Display top 15 most frequent words (and bigrams)
print("Top 50 Most Frequent Words in the Entire Document (Excluding Stop Words):")
for term, freq in sorted_frequency[:50]:
    print(f"{term}: {freq}")
```
